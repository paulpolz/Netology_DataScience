```python
names = ['name','landmass','zone','area','population','language','religion','bars','stripes',
         'colours','red','green','blue','gold','white','black','orange','mainhue','circles',
         'crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate',
         'text','topleft','botright']
```

```python
data = pd.read_table('flag.data', sep=',', names=names)
data.head()
```

```python
def color_negative_red(val):
    color = 'red' if val < - 0.6 else 'black'
    #color = 'blue' if val > 0.6 else 'black'
    return 'color: %s' % color
```

## Работа с категориальными данными

```python
categorical_columns = [i for i in data.columns if data[i].dtype.name == 'object']
numerical_columns = [i for i in data.columns if data[i].dtype.name != 'object']
```

```python
categorical_columns
```

```python
#Все категориальные признаки небинарные; используем для них метод get dummies
data[['name', 'mainhue', 'topleft', 'botright']].describe()
```

```python
categorical_data = pd.get_dummies(data[categorical_columns])
# удаляем неообработанные версии данных
data = data.drop(columns=data[categorical_columns])
```

```python
data.head()
```

```python
# У переменной area сильно отличается размерность. 
# Также у нее практически отсутсвует корелция со всеми переменными, кроме population
# AREA можно исключить для наилучщей нормализации данных
data = data.drop(columns=data[['area']])
```

```python
data.head()
```

```python
# Объединение данных
data = pd.concat((data, categorical_data), axis=1)

y = data[['religion']]
x = data.drop(columns=data[['religion']]) # Выбрасываем целевую переменную
```

```python
print(y.shape, x.shape)
```

## Подготовка обучающей и тестовой выборок

```python
X_test, y_test, X_train, y_train = train_test_split(x, y, test_size=0.2)
print(X_test.shape, y_test.shape, X_train.shape, y_train.shape)
```
